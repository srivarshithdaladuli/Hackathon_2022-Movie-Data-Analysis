```python
hack = pd.read_csv(r"C:\Users\SRIVARSHITH DALADULI\Downloads/data.csv")
hack.head()
```

```python
hack.drop(columns=['Movie Info','Unnamed: 0'],inplace=True,axis=1)
hack.head()
```

```python
hack.info()
```

```python
hack.isnull().sum()
```

```python
hack['License'].fillna('PG-13',inplace=True)
hack.info()
```

```python
hack1=hack.dropna(axis=0)
hack1.head()
```

```python
hack1['Release Date'] = pd.to_datetime(hack1['Release Date'],dayfirst = False,yearfirst= False)
hack1['Movie Runtime'] = hack1['Movie Runtime'].apply(lambda x:x.replace('hr',""))
hack1['Movie Runtime'] = hack1['Movie Runtime'].apply(lambda x:x.replace('min',""))
hack1.head()
```

```python
hack1.describe()
```

```python
hack1.info()
```

```python
l=[]
for i in hack1['Movie Runtime']: 
    l.append((int(i.split()[0])*60))
```

```python
hack1['Movie Runtime1'] = l
```

```python
hack1
```

```python
m=[]
for i in hack1['Movie Runtime']: 
    m.append(i.replace(i.split()[0],''))
```

```python
hack1['Movie Runtime2'] = m
```

```python
hack1.head()
```

```python
hack1['Movie Runtime2'] = hack1['Movie Runtime2'].apply(lambda x:x.replace(' ',""))
```

```python
hack1['Movie Runtime2'].unique()
```

```python
n=[]
for i in hack1['Movie Runtime2']:
    if i=="":
        n.append(0)
    else:
        n.append(i)
```

```python
hack1['Movie Runtime3'] = n
```

```python
hack.head()
```

```python
hack1['Movie Runtime3'].unique()
```

```python
hack1['Movie Runtime3'] = hack1['Movie Runtime3'].astype(str).astype('int64')
```

```python
hack1['Movie Runtime4'] = hack1['Movie Runtime1']+hack1['Movie Runtime3']
```

```python
hack1.head()
```

```python
hack1.info()
```

```python
hack1.drop(columns=['Movie Runtime1','Movie Runtime2','Movie Runtime3','Movie Runtime'],inplace=True,axis=1)
```

```python
hack1.head()
```

```python
hack1.info()
```

```python
hack1['Release Date'] = hack1['Release Date'].astype('datetime64[ns]').astype('str')
hack1['Release Year'] = hack1['Release Date'].apply(lambda x:re.search('\d+',x).group())
hack1.head()
```

```python
hack1['Genre'] = hack1['Genre'].apply(lambda x:x.replace('[',""))
```

```python
hack1['Genre'] = hack1['Genre'].apply(lambda x:x.replace(']',""))
```

```python
hack1['Genre'] = hack1['Genre'].apply(lambda x:x.replace(',',""))
```

```python
hack1.head()
```

```python
hack1.info()
```

```python
hack1['Domestic Sales (in $)'] = hack1['Domestic Sales (in $)'].astype(float)
```

```python
hack1['International Sales (in $)'] = hack1['International Sales (in $)'].astype(float)
```

```python
hack1['World Sales (in $)'] = hack1['World Sales (in $)'].astype(float)
```

```python
hack1.head()
```

```python
hack1.describe()
```

```python
hack1.drop(columns=['Release Date'],inplace=True,axis=1)
```

```python
hack1.rename(columns={'Movie Runtime4':'Movie Runtime','Release Year':'Movie Releases Year'},inplace=True)
```
